```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib
```

```python
df_a2c = pd.read_csv('df_daily_return_2.csv')
```

```python
df_dji = pd.read_csv('df_baseline_data.csv')
```

```python
df_dji_plot = df_dji[df_dji.date <= '2020-11-30']
```

```python
df_dji_plot['return'] = df_dji_plot['close']/df_dji_plot['close'][0]*100
```

```python
df_dji_plot['close'][0]
```

```python
a2c_cumpod =(df_a2c.daily_return+1).cumprod()*100
```

```python
# xtic=['2019-01-02','2019-04-01','2019-07-01']
xtic=[0,1,2]
```

```python
plt.rcParams["font.size"] = 16 #フォントサイズの指定
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111,ylabel='累積リターン [%]')
ax.plot(df_a2c.date,a2c_cumpod)
ax.plot(df_dji_plot.date,df_dji_plot['return'])
ax.set_xticks([0,124,252,377,482])
leg = ax.legend(['強化学習','インデックス'])
# ax.set_xticklabels(rotation=45)
fig.savefig('return.png')
```

```python
df_ac = pd.read_csv('df_actions_2.csv')
```

```python
df_ac_mean = df_ac.mean('index')
```

```python
df_ac_mean_s = df_ac_mean.sort_values()
```

```python
tickes = ['INTC','JNJ','CRM','AMGN','IBM']
# tickes = ['INTC']
```

```python
df_ac_1 = df_ac[df_ac.date <= '2019-02-01']
```

```python
df_ac_1[tickes] = df_ac_1[tickes]*100
```

```python
plt.rcParams["font.size"] = 16 #フォントサイズの指定
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111,ylabel='構成比率 [%]')
ax.stackplot(df_ac_1.date,df_ac_1['INTC'],df_ac_1['JNJ'],df_ac_1['CRM'],df_ac_1['AMGN'],df_ac_1['IBM'])
ax.set_xticks([0,21])
leg = ax.legend(tickes)
# ax.set_xticklabels(rotation=45)
fig.savefig('port.png')
```
